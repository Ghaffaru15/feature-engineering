```python
import pandas as pd
import numpy as np
import matplotlib
import datetime
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
```

```python
register_matplotlib_converters()
```

```python
apple_data = pd.read_csv('apple.csv')
```

```python
apple_data.head()
```

```python
apple_data = apple_data.sample(frac=1)
apple_data.reset_index(drop = True, inplace=True)
apple_data.head()
```

```python
apple_data['Date'] = pd.to_datetime(apple_data['Date'])
```

```python
apple_data.info()
```

```python
apple_data.shape
```

```python
apple_data.sort_values(by = 'Date').head()
```

```python
apple_data.head()
```

```python
apple_data['Date'] = apple_data['Date'].apply(lambda d: d.to_pydatetime())
```

```python
apple_data['Weekday'] = apple_data['Date'].apply(lambda d: d.isoweekday())
```

```python
apple_data
```

```python
extract_year = lambda x: datetime.datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S').year
```

```python
apple_data['Year'] = apple_data['Date'].apply(extract_year)
```

```python
extract_month = lambda x: datetime.datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S').month
```

```python
apple_data['Month'] = apple_data['Date'].apply(extract_month)
```

```python
data_by_month = apple_data.groupby(by = 'Month').mean()
data_by_month
```
