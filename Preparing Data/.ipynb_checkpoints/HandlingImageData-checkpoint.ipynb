```python
image = coins()
plt.figure(figsize=(10, 8))
plt.imshow(image, cmap='gray')
```

### Feature Detection

```python
edge = sobel(image)
```

```python
plt.figure(figsize=(10,8))
io.imshow(edge)
```

```python
cordinates = peak_local_max(image, min_distance=20)

fig, axes = plt.subplots(figsize=(12, 6))
axes.imshow(image, cmap = plt.cm.gray)

axes.plot(cordinates[:, 1], cordinates[:,0], 'r*')
plt.show()
```
